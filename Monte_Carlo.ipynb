#Monte Carlo
El método de Monte Carlo consiste en utilizar números aleatorios dentro de una superficie delimitada, es decir, para una función en la cual se desea encontrar una integral, se delimita con un rectángulo y se obtiene una proporcionalidad:
\begin{equation}
\frac{A_f}{A_r}=\frac {n(p)}{n(P)}
\end{equation}
donde $A_f$ es el área debajo de la curva $f(x)$, $A_r$ es el área debajo del rectángulo, $n(p)$ es el número de puntos debajo de la curva y $n(P)$ es el número de puntos totales.

```python
#Código hecho por: Diana Ailed Hernández Bustos

import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt
import random
from random import uniform
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import style
from math import*

#Definiendo las funciones de 2 dimensiones en adelante
def f3(x,y,R):
    if (R**2-x**2-y**2) > 0:       #Esfera
        return sqrt(R**2-x**2-y**2)
    else:
        return 0.0

def f4(x,y,z, R):
    if (R**2-x**2-y**2 - z**2) > 0:       #Esfera
        return sqrt(R**2-x**2-y**2- z**2)
    else:
        return 0.0

def f5(x,y, z, w, R):
    if (R**2-x**2-y**2 - z**2 - w**2) > 0:       #Esfera
        return sqrt(R**2-x**2-y**2- z**2 - w**2)
    else:
        return 0.0

def f6(x,y, z, w, u, R):
    if (R**2-x**2-y**2 - z**2 - w**2 -u**2) > 0:       #Esfera
        return sqrt(R**2-x**2-y**2- z**2 - w**2 - u**2)
    else:
        return 0.0

def f7(x,y, z, w, u, l, R):
    if (R**2-x**2-y**2 - z**2 - w**2 -u**2 - l**2) > 0:       #Esfera
        return sqrt(R**2-x**2-y**2- z**2 - w**2 - u**2-l**2)
    else:
        return 0.0

def f8(x,y, z, w, u, l, s,  R):
    if (R**2-x**2-y**2 - z**2 - w**2 -u**2 - l**2 -s**2) > 0:       #Esfera
        return sqrt(R**2-x**2-y**2- z**2 - w**2 - u**2-l**2 -s**2)
    else:
        return 0.0

def evol(NT, n):
    np = 0
    for i in range (1, NT):
        x=uniform(x0,x1)
        y=uniform(y0,y1)
        z=uniform(z0,z1)
        w = uniform(w0, w1)
        u = uniform(u0, u1)
        l=uniform(l0,l1)
        s = uniform(s0, s1)
        j = uniform(j0, j1)

        if n == 3:
            if z < f3(x,y,1):
                np=np+1.
                xprint.append(x)
                yprint.append(y)
                zprint.append(z)

        if n == 4:
            if w < f4(x,y,z, 1):
                np=np+1.
                xprint.append(x)
                yprint.append(y)
                zprint.append(z)
                wprint.append(w)

        if n == 5:
            if u < f5(x,y,z,w, 1):
                np=np+1.
                xprint.append(x)
                yprint.append(y)
                zprint.append(z)
                wprint.append(w)
                uprint.append(u)

        if n == 6:
            if l < f6(x,y,z,w, u, 1):
                np=np+1.
                xprint.append(x)
                yprint.append(y)
                zprint.append(z)
                wprint.append(w)
                uprint.append(u)

        if n == 7:
            if s < f7(x,y,z,w,u, l, 1):
                np=np+1.
                xprint.append(x)
                yprint.append(y)
                zprint.append(z)
                wprint.append(w)
                uprint.append(u)

        if n == 8:
            if j < f8(x,y,z,w,u,l,s, 1):
                np=np+1.
                xprint.append(x)
                yprint.append(y)
                zprint.append(z)
                wprint.append(w)
                uprint.append(u)
    return np
```
